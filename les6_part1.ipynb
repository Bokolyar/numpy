# Задание 1

```python
import warnings

warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
data = boston["data"]
feature_names = boston["feature_names"]
target = boston["target"]
```

```python
X = pd.DataFrame(data, columns=feature_names)
y = pd.DataFrame(target, columns=["price"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head(10)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(check_test["y_test"], check_test["y_pred"])
```

# Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(max_depth=12, n_estimators=1000, random_state=42)
model.fit(X_train, y_train.values[:, 0])
y_pred = model.predict(X_test)
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head(10)
```

```python
r2_score(check_test["y_test"], check_test["y_pred"])
```

##### Т.к. в задании 2 оценка r2_score ближе к 1, чем в задании 1, то 2-й метод моделирования оказался более точен.
